```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('recipes.csv')
```

```python
df
```

```python
df.describe()
```

```python
# Number of countries
print('List of Countries: ',df['country'].unique())
print('Number of Countries: ',len(df['country'].unique()))
print(df['country'].value_counts().reset_index())
```

```python
# Removing duplicated countries
df['country'] = df['country'].replace({
    'asian':'Asian',
    'east_asian':'Asian',
    'East-African':'African',
    'North-African':'African',
    'South-African':'African',
    'West-African':'African',
    'China':'Chinese',
    'chinese':'Chinese',
    'France':'French',
    'Germany':'German',
    'India':'Indian',
    'Italy':'Italian',
    'italian':'Italian',
    'Japan':'Japanese',
    'japanese':'Japanese',
    'korean':'Korean',
    'Korea':'Korean',
    'Mexico':'Mexican',
    'mexico':'Mexican',
    'Thailand':'Thai',
    'Vietnam':'Vietnamese'})
```

```python
# Number of countries after cleaning
print('List of Countries: ',df['country'].unique())
print('Number of Countries: ',len(df['country'].unique()))
print(df['country'].value_counts().reset_index())
```

```python
df.isnull().sum()
```

```python
# Countries that have more than 1000 recipes
countries = df['country'].value_counts().reset_index()
countries = countries[countries['count'] > 1000]
countries
```

```python
# Plot of the countries that have more than 1000 recipes
plt.figure(figsize=[10,6])
plt.pie(data=countries,x='count', autopct='%1.1f%%')
plt.legend(labels=countries['country'], loc='upper left')
plt.axis('equal')
plt.show()
```

```python
# Dataframe with countries that have more than 1000 recipes
filter_countries = df[df['country'].isin(countries['country'].tolist())]
filter_countries
```

```python
# Top 10 Countries
df_top_10 = df['country'].value_counts()[0:10].reset_index()
df_top_10
```

```python
df['country'].value_counts().reset_index().iloc[0:10,0:2]
```

```python
# Pie of TOP 10 Countries
plt.figure(figsize=[10,6])
plt.pie(data=df_top_10,x='count', autopct='%1.1f%%')
plt.legend(labels=df_top_10['country'], loc='upper left')
plt.axis('equal')
plt.show()
```
